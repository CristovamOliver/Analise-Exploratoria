# 5 -  Qual segmento teve o maior total de vendas ?

```python
filtro = dados.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda', 
                                                                                  ascending = False)
```

```python
filtro.head()
```

```python
# converter os dados de Valor_Venda em valor absoluto
```

```python
# 1 - minha conversão.
# absoluto = filtro[['Valor_Venda']].astype(int)
```

```python
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return '$ {v:d}'.format(v = val)
    return my_format
```

```python
plt.figure(figsize = (16, 6))


# Gráfico de pizza
plt.pie(filtro['Valor_Venda'], 
        labels = filtro['Segmento'], 
        autopct = autopct_format(filtro['Valor_Venda']), 
        startangle = 90)


#Limpa o círculo central
circulo = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(circulo)


#Labels e anotações
plt.annotate(text = 'Total de vendas: ' + '$ ' + str(int(sum(filtro['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de vendas por segmento')
plt.show()
```

```python
media_com_desconto = vendas_com_desconto.mean()
```

# 9 - Qual a média de vendas por segmento, por ano e por mês ? 

```python
# Extraímos o mês e gravamos em uma nova variável
dados['Mes'] = dados['Data_Pedido'].dt.month
```

```python
dados.head()
```

```python
# Agrupando por ano, mês e segmento e calculando a média e EXTRA(soma e mediana) .
agrupamento = dados.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
agrupamento
```

```python
# Pegando índices
ano = agrupamento.index.get_level_values(0)
mes = agrupamento.index.get_level_values(1)
segmento = agrupamento.index.get_level_values(2)
```

```python
plt.figure(figsize = (12, 6))
sns.set()
figura = sns.relplot(kind = 'line',
                    data = agrupamento,
                    y = 'mean',
                    x = mes,
                    hue = segmento,
                    col = ano,
                    col_wrap = 4)
plt.show()
```

# 10 - Qual o total de vendas por categoria e subcategoria, considerando somente as top 12 subcategorias ?

```python
agrupamento = dados.groupby(['Categoria', 'SubCategoria']).sum(numeric_only = True).sort_values(by = 'Valor_Venda', ascending = False).head(12)
```

```python
valor_absoluto = agrupamento[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
valor_absoluto
```

```python
# Criando outro dataframe para validar o valor_venda total por categoria.
```

```python
total = valor_absoluto.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
total
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                   '#0ee84f',
                   '#e80e27']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                      '#aa8cd5',
                      '#aa8cd6',
                      '#aa8cd7',
                      '#26c957',
                      '#26c958',
                      '#26c959',
                      '#26c960',
                      '#e65e65',
                      '#e65e66',
                      '#e65e67',
                      '#e65e68']
```

```python
#Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18, 12))

# Gráfico das categorias
p1 = ax.pie(total['Valor_Venda'],
           radius = 1,
           labels = total['Categoria'],
           wedgeprops = dict(edgecolor = 'white'),
           colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(valor_absoluto['Valor_Venda'],
           radius = 0.9,
           labels = valor_absoluto['SubCategoria'],
           autopct = autopct_format(valor_absoluto['Valor_Venda']),
           colors = cores_subcategorias,
           labeldistance = 0.7,
           wedgeprops = dict(edgecolor = 'white'),
           pctdistance = 0.53,
           rotatelabels = True
           )

# Limpa o cento do círculo
center_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(valor_absoluto['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de vendas por categoria e top 12 subcategorias')
plt.show()

```
